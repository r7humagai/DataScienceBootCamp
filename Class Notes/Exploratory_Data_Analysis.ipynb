      Pie Chart

```python
# Calculates the count of unique values in the 'Country' column of the 'final_df' dataFrame
# index retrieves the index(the unique country code names) of the resulting series
# Variable 'country_names' gets: list of unique country names of 'Country' column

country_names = final_df.Country.value_counts().index
print(type(country_names))
```

```python
# Calculates the count of unique values in the 'Country' of the final_df DataFrame
#. values retrieves the count of the resulting series
#. 'country_val' variable gets: list of counts corresponding to each unique country name -> type: numpy array
country_val = final_df.Country.value_counts().values
print(type(country_val))
```

```python
# plt.pie -> used to create a pie chart -> function from Matplotlib
# country_val[:4] -> Only take the top 3 values to be plotted in the pie chart [Top 3 countries]
# labels = country_names[:3] -> specifies the labels for the pie chart segments. [:3]-> First three countries
# autopct='%1.2f%%' -> (parameter) specifies format for automatic percentage
#                   -> formats the percentages with 2 decimal places
```

```python
plt.pie(country_val[:4],labels = country_names[:4],autopct = '%2.2f%%')
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12, 6)
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings)
```

```python
sns.barplot(x="Aggregate rating",y="Rating Count",hue='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','green'])
```

```python
## Count plot
sns.countplot(x="Rating color",data=ratings,palette=['blue','red','orange','yellow','green','green'])
```

```python
final_df.City.value_counts().index
city_values=final_df.City.value_counts().values
city_labels=final_df.City.value_counts().index
plt.pie(city_values[:5],labels=city_labels[:5],autopct='%2.2f%%')
```
