```python
import pandas as pd
import numpy as np
```

```python
path = '/content/Ch1_Q21_V09_Data_File.xlsx'
df_q2_raw = pd.read_excel(path)
df_q2_raw
```

```python
# Cleaning the data:
'Take necessary columns'
df_q2_clean = df_q2_raw.iloc[0: ,0:5]
df_q2_clean

'Check for null values'
print(df_q2_clean.isna().sum())

'Delete the rows with null values'
df_q2_final = df_q2_clean.dropna()
df_q2_final
```

```python
'''
Question a.
How many observations have x4 values that are less than 3?
'''

answer = (df_q2_final['x4'] <3).sum()
print(f"{answer} observations have x4 values that are less than 3.")
```

```python
'''
Question b
Sort the data by x1 in ascending order.
What are the x1, x2, x3, and x4 values of the first observation after the data are sorted?
(For x1, round your answer to four decimal places.)
'''

df_ascending = df_q2_final.sort_values(['x1'], ascending = True)
df_ascending['x1'] = df_ascending['x1'].apply(lambda x:f"{x:.4f}")
df_ascending.iloc[0]
```

```python
'''
Question c
Sort the data by x1 in descending order.
What are the x1, x2, x3, and x4 values of the first observation after the data are sorted?
(For x1, round your answer to four decimal places.)
'''
df_descending = df_q2_final.sort_values(['x1'], ascending = False)
df_descending['x1'] = df_descending['x1'].apply(lambda x:f"{x:.4f}")
df_descending.iloc[0]
```

```python
'''
Question d
How many missing values are there in x1, x2, x3, and x4?
'''

print(df_q2_clean.isna().sum())
```

```python
'''
How many observations are there in each category in x4?
'''

df_cat_x4 = df_q2_final['x4'].value_counts().reset_index()
df_cat_x4.columns = ['x4', 'Observations']

print(df_cat_x4)
```
