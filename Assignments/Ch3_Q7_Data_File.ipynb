```python
import pandas as pd
import numpy as np
```

```python
path = '/content/Ch3_Q7_Data_File.xlsx'
```

```python
df_raw = pd.read_excel(path)
```

```python
df_raw.head(10)
```

```python
df_clean = df_raw.iloc[0:,0:3]
df_clean.shape
```

```python
df_clean.head(10)
```

```python
'Find the mean and median life expectancy. (Round your answers to 2 decimal places.)'
# Calculating Mean

total_age = 0

for age in df_clean['Life_Expectancy']:
  total_age += age

mean = total_age / df_clean.shape[0]
mean = '{:.2f}'.format(mean)
print("Mean Age:", mean,"years")

```

```python
# Calculating Median age life expectancy

# Sort the sale prices in ascending order
sorted_age = df_clean['Life_Expectancy'].sort_values().reset_index(drop=True)
# Calculate the median
n = df_clean.shape[0]

if n % 2 == 0:
    # If there's an even number of values, take the average of the two middle values
    middle1 = sorted_age[n // 2 - 1]
    middle2 = sorted_age[n // 2]
    median = (middle1 + middle2) / 2
else:
    # If there's an odd number of values, take the middle value
    median = sorted_age[n // 2]

median = '{:.2f}'.format(median)
print("Median Age:", median,"years")
```
