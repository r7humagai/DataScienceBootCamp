```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
path = '/content/Ch03_Q71_V03_Data-File.xlsx'
```

```python
df_raw = pd.read_excel(path)
```

```python
df_raw.head()
```

```python
# Cleaning the data: Filter the necesssary columns
df_clean = df_raw.iloc[0: ,0:2]
df_clean.isnull().sum()

# Remove the null values

df_clean = df_clean.dropna()
```

```python
df_clean
```

```python
'''
a. Calculate the covariance between the variables.
(Negative value should be indicated by a minus sign.
Round your intermediate calculations to at least 4 decimal places and final answer to 2 decimal places.)
'''
# Calculating the mean:
meanX = '{:.4f}'.format(df_clean['x'].mean())
meanY = '{:.4f}'.format(df_clean['y'].mean())
# Deviations:

dev_x = df_clean['x'].astype(float) - float(meanX)
dev_y = df_clean['y'].astype(float) - float(meanY)

# Covariance:
covariance = (dev_x * dev_y).sum() / (len(df_clean) - 1)

covariance = '{:.2f}'.format(covariance)

print("Covariance between X and Y:", covariance)
```

```python
'''
b. Calculate the correlation coefficient.
(Round your intermediate calculations to 4 decimal places and final answer to 2 decimal places.)
'''
# Standard Deviation
std_dev_x = '{:.4f}'.format(df_clean['x'].std())
std_dev_y = '{:.4f}'.format(df_clean['y'].std())

# correlation coefficient
correlation = -31.75 / (float(std_dev_x) * float(std_dev_y))

correlation = '{:.2f}'.format(correlation)

print("Correlation coefficient between 'x' and 'y':", correlation)
```
