```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
path = '/content/Ch2_Q57_Data_File.xlsx'
```

```python
df_raw = pd.read_excel(path)
```

```python
df_raw.head(10)
df_raw.shape
```

```python
# Cleaning the data: Only keeping the relevant columns
df_clean = df_raw.iloc[0: ,0:4]
df_clean.head(5)
```

```python
'''
The manager of a human resources department of a large manufacturing firm gathers data on
the hourly wage (in $), years of education, and age for 80 employees. The accompanying table shows a portion of the data.
'''
```

```python
# Question a: Construct a scatterplot of Wage and Education.

plt.style.use('seaborn')

wage = df_clean['Wage']
education = df_clean['Education']
wage_per_education = wage/education

plt.scatter(education, wage, c=wage_per_education, cmap = 'summer', edgecolor = 'black', linewidth = 1, alpha = 0.75)

cbar = plt.colorbar()
cbar.set_label('Wage/Education Ratio')

plt.title('Wage-Education Ratio')
plt.xlabel('Education')
plt.ylabel('Wage')

plt.show()
```

```python
'''
Question a: Construct a scatterplot of Wage and Education.
Given the scatterplot, which statement best describes the relationship between Wage and Education?
i. Wage and Education have a positive relationship.
ii. Wage and Education have a negative relationship.
iii. Wage and Education have no relationship.
iv. Wage and Education have an inverted U-shaped (nonlinear) relationship.
'''

print("Wage and Education have a positive relation as we can see from the scatterplot, the wage increases as education increases.")
```

```python
# b. Construct a scatterplot of Wage and Age

plt.style.use('seaborn')

wage = df_clean['Wage']
age = df_clean['Age']
wage_per_age = wage/age

plt.scatter(age, wage, c=wage_per_age, cmap = 'summer', edgecolor = 'black', linewidth = 1, alpha = 0.75)

cbar = plt.colorbar()
cbar.set_label('Wage/Age Ratio')

plt.title('Wage-Age Ratio')
plt.xlabel('Age')
plt.ylabel('Wage')

plt.show()
```
