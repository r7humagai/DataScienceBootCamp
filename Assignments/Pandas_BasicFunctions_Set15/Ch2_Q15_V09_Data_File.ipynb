```python
import pandas as pd
import numpy as np
```

```python
path = '/content/Ch2_Q15_V09_Data_File.xlsx'
```

```python
df_q6_raw = pd.read_excel(path)
```

```python
df_q6_raw.head(10)
df_q6_raw.shape
```

```python
# Cleaning the data
df_q6_clean = df_q6_raw.iloc[0:,0:3]
print(df_q6_clean)

# Check for null values
df_q6_clean.isna().sum()    # No null values
```

```python
df_q6 = df_q6_clean
df_q6
```

```python
'a-1.Â Construct a contingency table that cross-classifies the data by Sex and Feasible. Provide the frequencies in the accompanying table.'
df_q6_grouping = df_q6.groupby(['Sex', 'Feasible'], as_index = True).count()
df_q6_grouping
```

```python
'a-2. How many of the students were female?'

number_of_female_students = (df_q6['Sex'] == 'female').sum()
print("The students that are female: ",number_of_female_students)
```

```python
'a-3.Â How many of the students felt that it was feasible for men and women to be just friends?'

answer3 = (df_q6['Feasible'] == "yes").sum()
print(answer3, "of the students felt that it was feasible for men and women to be just friends.")
```

```python
'''
b-1.Â What is the likelihood that a male student feels that men and women can be just friends?
(Report the proportion rounded to 2 decimal places.)
'''
# Calculate the number of male students
number_of_male_students = (df_q6['Sex'] == 'male').sum()

# Calculate the number of students who are Male & Feasible is yes

answerb1 = ((df_q6['Sex'] == 'male') & (df_q6['Feasible'] == 'yes')).sum()

# Calculate a percentage value
likelihood = "{:.2f}".format((answerb1/number_of_male_students) * 100)
print(likelihood, "students of 100 male students feel that men and women can be just friends.")
```

```python
'''
b-2.Â What is the likelihood that a female student feels that men and women can be just friends?
(Report the proportion rounded to 2 decimal places.)
'''
# Calculate the number of female students
number_of_female_students = (df_q6['Sex'] == 'female').sum()

# Calculate the number of students who are Female & Feasible is yes

answerb2 = ((df_q6['Sex'] == 'female') & (df_q6['Feasible'] == 'yes')).sum()

# Calculate a percentage value
likelihood2 = "{:.2f}".format((answerb2/number_of_female_students) * 100)
print(likelihood2, "students of 100 female students feel that men and women can be just friends.")
```
