```python
import pandas as pd
import numpy as np
```

```python
path = '/content/Ch1_Q20_V12_Data_File.xlsx'
```

```python
df_q1_raw = pd.read_excel(path)
```

```python
df_q1 = df_q1_raw.iloc[ : ,0:4]
df_q1
```

```python
# Cleaning the data

'Step 1: Check if there are any entry with null values'

print(df_q1.isna().sum())

'Since there are datas with null values, we will remove them'

df_q1_final = df_q1.dropna()
print(df_q1_final)
```

```python
# Question a
'a: How many observations have x1 values greater than 30?'

answer = (df_q1_final.iloc[:,1:2] > 30).sum()
print(answer)
```

```python
# Question b
'b. Sort the data by x1 in ascending order. What are the x1, x2, and x3 values of the first observation after the data are sorted?'

df_ascending = df_q1_final.sort_values(by = 'x1', ascending = True)

print(df_ascending.iloc[0,1:4], "\n")
df_ascending
```

```python
# Question c
'c. Sort the data by x1 in descending order. What are the x1, x2, and x3 values of the first observation after the data are sorted?'

df_descending = df_q1_final.sort_values(by = 'x1', ascending = False)

print(df_descending.iloc[0,1:4], "\n")
df_descending
```

```python
# Question d
'd. How many missing values are there in x1, x2, and x3?'

print(df_q1[['x1','x2','x3']].isna().sum())
```
