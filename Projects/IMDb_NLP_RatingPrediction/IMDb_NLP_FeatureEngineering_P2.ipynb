#**NLP Implementation on 'overview', 'plot_keyword'**

```python
# overview
df['overview'] = df['overview'].str.lower()
df['plot_keyword'] = df['plot_keyword'].str.lower()
```

```python
# Remove Stop Words

import nltk
nltk.download('stopwords')
nltk.download('punkt')

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

stop_words = set(stopwords.words("english"))

def remove_stopwords(text):
  words = word_tokenize(text)
  return ' '.join([word for word in words if word.lower() not in stop_words])

df['overview'] = df['overview'].apply(remove_stopwords)
df['plot_keyword'] = df['plot_keyword'].apply(remove_stopwords)
```

```python
print(df.overview[10:15])
print()
print(df.plot_keyword[10:15])
```

```python
# Remove Punctuation

from nltk.tokenize import RegexpTokenizer

def remove_punctuation(text):
  tokenizer = RegexpTokenizer(r'\w+')
  tokens = tokenizer.tokenize(text)
  return ' '.join(tokens)

df['overview'] = df['overview'].apply(remove_punctuation)
df['plot_keyword'] = df['plot_keyword'].apply(remove_punctuation)
```

```python
print(df.overview[10:15])
print()
print(df.plot_keyword[10:15])
```

```python
# Apply Stemming

from nltk.stem import PorterStemmer
stemmer = PorterStemmer()

def func_stemming(sentence):
  return ' '.join([stemmer.stem(word) for word in sentence.split()])

df['overview'] = df['overview'].apply(func_stemming)
df['plot_keyword'] = df['plot_keyword'].apply(func_stemming)
```
